```python
def get_eof(data, n_modes=3):
    model = xeofs.models.EOF(n_modes=n_modes, use_coslat=True)
    model.fit(data, dim='time')
    components = model.components().persist()
    variance_explained = model.explained_variance_ratio().persist()*100
    timeseries = model.fit_transform(data, dim='time', normalized=False).persist()
    del components.attrs['solver_kwargs']
    del variance_explained.attrs['solver_kwargs']
    return xr.Dataset(dict(components=components, 
                           timeseries=timeseries, 
                           variance_explained=variance_explained))

def get_eof_gf(data, gf, n_modes=3):
    eof = get_eof(data*gf, n_modes=n_modes)
    eof['components_sst'] = eof.components/gf
    eof['pe'] = (eof.components*eof.timeseries).sum(['lon','lat'])
    eof['pe_total'] = (data*gf).sum(['lon','lat'])
    eof['variance_explained_pe'] = eof.pe.var('time')/eof.pe_total.var('time')*100
    return eof
```

# Preprocessing
Before this analysis, both SST data and Green's Functions are regridded to a 1degx1deg grid. 
SST are regridded using bilinear interpolation and GF are regridded using a conservative regridding using `xesmf`.
For this notebook, the ocean basin mask is in `OCEAN_MASK_PATH`.

```python
# Load SST product
RAW_SST_PATH = '/projets/EMC2/data/Observations/SST/up_to_2025' # From products webpages directly

ds_hadisst = xr.open_dataset(f'{RAW_SST_PATH}/HadISST_1870-2025.nc').sst
ds_cobe = xr.open_dataset(f'{RAW_SST_PATH}/COBE2_1850-2025.nc').sst
ds_ersst = xr.open_dataset(f'{RAW_SST_PATH}/ERSSTv5_1854-2025.nc').sst

# Align time axis
ds_cobe = ds_cobe.sel(time=slice('1870','2024'))
ds_ersst = ds_ersst.sel(time=slice('1870','2024')).assign_coords(time=ds_cobe.time)
ds_hadisst = ds_hadisst.sel(time=slice('1870','2024')).assign_coords(time=ds_cobe.time)

# Regrid to a 1deg x 1deg grid
out_grid = xr.DataArray(dims=['longitude','latitude'],
                        coords=dict(longitude=np.arange(1,360,2), latitude=np.arange(89,-90,-2)))

hadisst_regridded = xe.Regridder(ds_hadisst, out_grid, 'conservative_normed', periodic=True)(ds_hadisst)
cobe_regridded = xe.Regridder(ds_cobe, out_grid, 'conservative_normed', periodic=True)(ds_cobe)
ersst_regridded = xe.Regridder(ds_ersst, out_grid, 'conservative_normed', periodic=True)(ds_ersst)

# Combine SST products in one dataset
SST = xr.concat([hadisst_regridded.assign_coords(product='HadISST'),
                cobe_regridded.assign_coords(product='COBE2'),
                ersst_regridded.assign_coords(product='ERSSTv5'),
                ], dim='product').rename('SST').rename(longitude='lon', latitude='lat')

```

```python

# Remove climatology and make anomalies
SST_climatology = SST.lntime.climato()
SST_ref = SST_climatology.sel(time=slice('1870', '1900')).mean('time')
SST_temporal_anom = SST_climatology - SST_ref
SST_mean_temporal_anom = SST_temporal_anom.weighted(np.cos(np.deg2rad(SST_temporal_anom.lat))).mean(['lon','lat'])
SSTA = SST_temporal_anom - SST_mean_temporal_anom
SSTA_60 = SSTA.sel(time=slice('1960',None))
```

```python
# Remove a quadratic trend to remove forced signal

SSTA_detrend = SSTA_60 - xr.polyval(SSTA_60.time, SSTA_60.polyfit('time', deg=2).polyfit_coefficients)
SSTA_detrend = SSTA_detrend.where(SSTA_detrend.count('time') == SSTA_detrend.count('time').max())
OCEAN_MASK_PATH = '/home/rguillaume/developpements/Publications/2025_Guillaume-Castel_Ceppi_PatternEffect-Variability/RECCAP2_region_masks_all_v20221025.nc'
ds_mask = xr.open_dataset(OCEAN_MASK_PATH).open_ocean
mask = ds_mask.interp_like(SSTA_detrend, method='nearest')
SSTA_pac = SSTA_detrend.where(mask==2, drop=True)
```

```python
# Load raw green's functions from preliminary datasets
RAW_GREENS_FUNCTIONS_PATH = '/projets/EMC2/data/Models/GFMIP/preliminary_spatial_feedbacks-main/data/' # From GFMIP preliminary dataset

ds_cam4 = xr.open_dataset(f'{RAW_GREENS_FUNCTIONS_PATH}cam4.nc').spatial_feedbacks_from_warming
ds_cam5 = xr.open_dataset(f'{RAW_GREENS_FUNCTIONS_PATH}cam5.nc').spatial_feedbacks_from_both
ds_canesm5 = xr.open_dataset(f'{RAW_GREENS_FUNCTIONS_PATH}canesm5.nc').spatial_feedbacks_from_both
ds_echam6 = xr.open_dataset(f'{RAW_GREENS_FUNCTIONS_PATH}echam6.nc').spatial_feedbacks_from_both
ds_gfdlam4 = xr.open_dataset(f'{RAW_GREENS_FUNCTIONS_PATH}gfdlam4.nc').spatial_feedbacks_from_both
ds_hadcm3 = xr.open_dataset(f'{RAW_GREENS_FUNCTIONS_PATH}hadcm3.nc').spatial_feedbacks_from_both

ds_cam4 = ds_cam4.assign_coords(longitude=ds_cam4.longitude.values, latitude=ds_cam4.latitude.values)
ds_cam4 = ds_cam4.where(ds_cam4!=0)
ds_cam5 = ds_cam5.assign_coords(longitude=ds_cam5.longitude.values, latitude=ds_cam5.latitude.values)
ds_cam5 = ds_cam5.where(ds_cam5!=0)
ds_canesm5 = ds_canesm5.assign_coords(longitude=ds_canesm5.longitude.values, latitude=ds_canesm5.latitude.values)
ds_canesm5 = ds_canesm5.where(ds_canesm5!=0)
ds_echam6 = ds_echam6.assign_coords(longitude=ds_echam6.longitude.values, latitude=ds_echam6.latitude.values)
ds_echam6 = ds_echam6.where(ds_echam6!=0)
ds_gfdlam4 = ds_gfdlam4.assign_coords(longitude=ds_gfdlam4.longitude.values, latitude=ds_gfdlam4.latitude.values)
ds_gfdlam4 = ds_gfdlam4.where(ds_gfdlam4!=0)
ds_hadcm3 = ds_hadcm3.assign_coords(longitude=ds_hadcm3.longitude.values, latitude=ds_hadcm3.latitude.values)
ds_hadcm3 = ds_hadcm3.where(ds_hadcm3!=0)


# Regrid to the same grid as sst
cam4_regridded = xe.Regridder(ds_cam4, out_grid, 'conservative_normed', periodic=True, )(ds_cam4)
cam5_regridded = xe.Regridder(ds_cam5, out_grid, 'conservative_normed', periodic=True)(ds_cam5)
canesm5_regridded = xe.Regridder(ds_canesm5, out_grid, 'conservative_normed', periodic=True)(ds_canesm5)
echam6_regridded = xe.Regridder(ds_echam6, out_grid, 'conservative_normed', periodic=True)(ds_echam6)
gfdlam4_regridded = xe.Regridder(ds_gfdlam4, out_grid, 'conservative_normed', periodic=True)(ds_gfdlam4)
hadcm3_regridded = xe.Regridder(ds_hadcm3, out_grid, 'conservative_normed', periodic=True)(ds_hadcm3)


# Combine green's functions into single dataarray
GF = xr.concat([cam4_regridded.assign_coords(GF_source='CAM4'),
                cam5_regridded.assign_coords(GF_source='CAM5'),
                canesm5_regridded.assign_coords(GF_source='CanESM5'),
                echam6_regridded.assign_coords(GF_source='ECHAM6'),
                gfdlam4_regridded.assign_coords(GF_source='GFDL-AM4'),
                hadcm3_regridded.assign_coords(GF_source='HadCM3')],
                dim='GF_source').rename('GF')
GF = GF.where(GF.count('GF_source')==GF.count('GF_source').max()).rename(longitude='lon', latitude='lat')
GF = (-GF/GF.sum(['lon','lat']))
GF = xr.concat([GF, GF.mean('GF_source').assign_coords(GF_source='mean')], dim='GF_source')
```

## Compute EOFs

```python
all_eofs = []
for product in tqdm(SSTA_detrend.product):
    all_eofs_product = []
    for GF_source in tqdm(GF.GF_source):
        eof_pac_gf = get_eof_gf(SSTA_detrend.sel(product=product), GF.sel(GF_source=GF_source),n_modes=20).assign_coords(GF_source=GF_source)
        all_eofs_product.append(eof_pac_gf)
        print(product.values, GF_source.values)
    eofs_pac_product = xr.concat(all_eofs_product, dim='GF_source').assign_coords(product=product)
    eofs_pac_product = eofs_pac_product.rename(dict(components="components",
                                    timeseries="timeseries",
                                    variance_explained="variance_explained",
                                    components_sst="components_sst",
                                    pe="pe",
                                    pe_total="pe_total_eofs_pac",
                                    variance_explained_pe="variance_explained_pe_eofs_pac",))
    all_eofs.append(eofs_pac_product)
all_eofs_pac = xr.concat(all_eofs, dim='product')



# Normalize all patters b sign and apmlitude
sign_eof = np.sign(all_eofs_pac.components_sst.sel(lon=200, lat=0, method='nearest'))
sign_eof.loc[dict(mode=2)] = -np.sign(all_eofs_pac.components_sst.sel(mode=2).where(all_eofs_pac.lat>50).mean(['lon','lat']))
sign_eof.loc[dict(mode=2, product='ERSSTv5')] = np.sign(all_eofs_pac.components_sst.sel(mode=2, product='ERSSTv5', lon =slice(245,250), lat=slice(-5,5)).mean(['lon','lat']))
sign_eofs =sign_eof#*(all_eofs_pac.timeseries.std('time')
lbd_timeseries = all_eofs_pac.timeseries.std('time')
lbd_pattern = all_eofs_pac.components_sst.weighted(np.cos(np.deg2rad(all_eofs_pac.lat))).std(['lon','lat'])

amplitude_eofs = sign_eof/all_eofs_pac.components_sst.std(['lon','lat'])
all_eofs_pac['timeseries'] = all_eofs_pac['timeseries']*sign_eof/lbd_timeseries
all_eofs_pac[['components_sst','components']] =all_eofs_pac[['components_sst','components']]*sign_eof/lbd_pattern
all_eofs_pac['lambdas'] = lbd_pattern*lbd_timeseries
all_eofs_pac['pi'] = (all_eofs_pac.components_sst*GF).sum(['lon','lat'])
all_eofs_pac = all_eofs_pac.rename(mode='feature')

all_eofs_pac[['components','components_sst','pi']] = all_eofs_pac[['components','components_sst','pi']]*np.sign(all_eofs_pac.pi)

eofs_pac = all_eofs_pac.isel(feature=np.arange(3))

eofs_pac['lambdas_scaled'] = eofs_pac.lambdas/Rp_pac.std('time')
```

## Compute PLSR

```python
all_output = []
for product in SSTA_detrend.product.values:
    pattern = SSTA_detrend.sel(product=product, drop=True)#.sel(time=slice('1960',None))
    all_product = []
    for gf_choice in tqdm(GF.GF_source):
        GF_mean = GF.sel(GF_source=gf_choice, drop=True)
        X = pattern#*GF_mean
        Y = Rp_pac.sel(GF_source=gf_choice, product=product).sel(time=pattern.time)
        X = X.sel(time=Y.time)

        weights_X = xr.ones_like(X)*np.cos(np.deg2rad(GF.lat))
        X = X.stack(space=['lon','lat'])
        weights_X = weights_X.stack(space=['lon','lat'])
        X = X.dropna('space')
        weights_X = weights_X.sel(space=X.space)
        
        n_components = 20
        pls2 = PLSRegression(n_components=n_components, scale=False)
        pls2.fit(X*weights_X, Y)

        components =  xr.DataArray(pls2.x_loadings_[:]/weights_X.isel(time=range(n_components), drop=True).T, dims=['space','mode'], coords=dict(space=X.space)).unstack().sortby(['lon','lat']).transpose('mode','lat','lon')        
        timeseries =  xr.DataArray(pls2.x_scores_, dims=['time','mode'], coords=dict(time=X.time))
        components_sst = components#/GF_mean
        components = components_sst*GF_mean
        ds_out_one = xr.Dataset(dict(components_sst=components_sst,
                                     components=components,
                                     timeseries=timeseries)).assign_coords(product=product, GF_source=gf_choice)
        ds_out_one['pe'] = (ds_out_one.components.sum(['lon','lat'])*ds_out_one.timeseries)
        all_product.append(ds_out_one)
        # plt.figure()
        # ds_out_one.pe.to_series().plot.barh
    ds_product = xr.concat(all_product, dim='GF_source').assign_coords(mode=np.arange(1,21))
    all_output.append(ds_product)
all_plsr_pac = xr.concat(all_output, dim='product')


# Normalize all patters b sign and apmlitude
sign_plsr = np.sign(all_plsr_pac.components_sst.sel(lon=200, lat=0, method='nearest'))
sign_plsr.loc[dict(mode=3)] = -np.sign(all_plsr_pac.components_sst.sel(mode=3).where(all_plsr_pac.lat>50).mean(['lon','lat']))

lbd_timeseries = all_plsr_pac.timeseries.std('time')
lbd_pattern = all_plsr_pac.components_sst.weighted(np.cos(np.deg2rad(all_plsr_pac.lat))).std(['lon','lat'])

all_plsr_pac['timeseries'] = all_plsr_pac['timeseries']*sign_plsr/lbd_timeseries
all_plsr_pac[['components_sst','components']] =all_plsr_pac[['components_sst','components']]*sign_plsr/lbd_pattern
all_plsr_pac['lambdas'] = lbd_pattern*lbd_timeseries
all_plsr_pac['pi'] = (all_plsr_pac.components_sst*GF).sum(['lon','lat'])
all_plsr_pac = all_plsr_pac.rename(mode='feature')
all_plsr_pac[['components','components_sst','pi']] = all_plsr_pac[['components','components_sst','pi']]*np.sign(all_plsr_pac.pi)

plsr_pac = all_plsr_pac.isel(feature=np.arange(3))
```

## Compute ENSO indices


```python

# EP CP ENSO from https://icar.nuist.edu.cn/en/NlNinoIndices/listm.psp
nino34_bounds = {"lat": slice(5, -5), "lon": slice(190, 240)}  # 160°E to 150°W
nino3_bounds = {"lat": slice(5, -5), "lon": slice(210, 270)}  # 150°W to 90°W
nino4_bounds = {"lat": slice(5, -5), "lon": slice(160, 210)}  # 160°E to 150°W

nino34 = SSTA.sel(nino34_bounds).weighted(np.cos(np.deg2rad(SSTA.lat))).mean(['lon','lat']) 
nino3 = SSTA.sel(nino3_bounds).weighted(np.cos(np.deg2rad(SSTA.lat))).mean(['lon','lat'])
nino4 = SSTA.sel(nino4_bounds).weighted(np.cos(np.deg2rad(SSTA.lat))).mean(['lon','lat'])
alpha = xr.ones_like(nino3).where(nino3*nino4>0,0)*2/5

EP = nino3 - alpha*nino4
CP = nino4 - alpha*nino3

eofs_trop_pac_SST = xr.concat([get_eof(SSTA_pac.sel(product=product).sel(lat=slice(10,-10))).timeseries \
                                for product in tqdm(SSTA_detrend.product)], dim='product')
nino34_eof = eofs_trop_pac_SST.isel(mode=0)
nino_modoki_eof = eofs_trop_pac_SST.isel(mode=1)
nino_C_takahashi = (nino34_eof+nino_modoki_eof)/np.sqrt(2)
nino_E_takahashi = (nino34_eof-nino_modoki_eof)/np.sqrt(2)

modes = xr.concat([nino34.assign_coords(mode='nino34'),
                   CP.assign_coords(mode='CP'),
                   EP.assign_coords(mode='EP'),
                   nino34_eof.assign_coords(mode='nino34_eof'),
                   nino_modoki_eof.assign_coords(mode='modoki_eof'),
                   nino_C_takahashi.assign_coords(mode='nino_C_takahashi'),
                   nino_E_takahashi.assign_coords(mode='nino_E_takahashi')
                   ], dim='mode')

# regr_modes = xr_linregress(modes, SSTA.where(mask==2, drop=True), dim='time')
regr_modes = xr_linregress(modes, SSTA_detrend, dim='time')
regr_modes['mode_timeseries'] = modes 
modes_obs = regr_modes.rename(slope='mode_pattern')[['mode_timeseries', 'mode_pattern']]

modes_obs['lambdas'] = modes_obs.mode_pattern.weighted(np.cos(np.deg2rad(modes_obs.lat))).std(['lon','lat'])
modes_obs['mode_pattern'] = modes_obs.mode_pattern/modes_obs.lambdas
modes_obs['pi'] = (modes_obs.mode_pattern*GF).sum(['lon','lat'])
```
